```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Load the dataset
data = pd.read_csv('Oder3.csv')

# Step 1: Generate list of transactions
transactions = data.groupby('TransactionNo')['Items'].apply(list).tolist()

# Step 2: Preprocess the transactions
te = TransactionEncoder()
te_data = te.fit_transform(transactions)
df_transactions = pd.DataFrame(te_data, columns=te.columns_)

# Step 3: Train Apriori
frequent_itemsets = apriori(df_transactions, min_support=0.01, use_colnames=True)

# Step 4: Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.2)
# rules = association_rules()
# Display results
print(frequent_itemsets.head())
print(rules.head())
```
