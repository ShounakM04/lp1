```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd

# Load data
df = pd.read_csv('Oder3.csv')

# Group items by transaction
df_transactions = df.groupby('TransactionNo')["Items"].apply(list).tolist()

# Apply TransactionEncoder
te = TransactionEncoder()
te_data = te.fit_transform(df_transactions)
df_transactions = pd.DataFrame(te_data, columns=te.columns_)

# Generate frequent itemsets with Apriori
list_items = apriori(df_transactions, min_support=0.02, use_colnames=True)

# Generate association rules
rules = association_rules(list_items, num_itemsets=len(list_items),metric='confidence', min_threshold=0.2)

# Filter rules with confidence greater than 0.25
good_rules = rules[rules['confidence'] > 0.25]

# Display results
print("Frequent Itemsets:")
print(list_items)
print("\nAssociation Rules (Confidence > 0.25):")
print(good_rules)
```
