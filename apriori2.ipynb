```python
import pandas as pd

df = pd.read_csv('Order2.csv')
df
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Step 1: Load the dataset (assuming df is your DataFrame)
# Convert each transaction row into a list of purchased items, dropping NaNs
transactions = df.dropna(axis=1).values.tolist()


# Step 2: Apply TransactionEncoder for one-hot encoding
te = TransactionEncoder()
te_data = te.fit_transform(transactions)

# Convert the encoded data into a DataFrame
df_transactions = pd.DataFrame(te_data, columns=te.columns_)

# Step 3: Apply Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(df_transactions, min_support=0.02, use_colnames=True)

# Step 4: Generate association rules
rules = association_rules(frequent_itemsets, num_itemsets=len(frequent_itemsets), metric="confidence", min_threshold=0.2)

# Display the frequent itemsets and rules
print("Frequent Itemsets:")
print(frequent_itemsets)
print("\nAssociation Rules:")
print(rules)
```
